```python
import cv2
import matplotlib.pyplot as plt

# Cargar imagen
img_color = cv2.imread("AlbedoBase_XL_a_highcontrast_cinematic_photo_of_a_character_co_1.jpg")  # cambia por el nombre real de tu imagen
img_color = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)

# Escala de grises
img_gray = cv2.cvtColor(img_color, cv2.COLOR_RGB2GRAY)

# Visualizacin
plt.subplot(1, 2, 1), plt.imshow(img_color), plt.title("Color"), plt.axis('off')
plt.subplot(1, 2, 2), plt.imshow(img_gray, cmap='gray'), plt.title("Grises"), plt.axis('off')
plt.show()
```

```python
import numpy as np

# Filtro blur
blurred = cv2.GaussianBlur(img_gray, (5, 5), 0)

# Filtro sharpening
kernel_sharp = np.array([[0, -1, 0],
                         [-1, 5, -1],
                         [0, -1, 0]])
sharpened = cv2.filter2D(img_gray, -1, kernel_sharp)

# Mostrar
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1), plt.imshow(img_gray, cmap='gray'), plt.title("Original"), plt.axis('off')
plt.subplot(1, 3, 2), plt.imshow(blurred, cmap='gray'), plt.title("Blur"), plt.axis('off')
plt.subplot(1, 3, 3), plt.imshow(sharpened, cmap='gray'), plt.title("Sharpen"), plt.axis('off')
plt.show()
```

```python
# Sobel
sobelx = cv2.Sobel(img_gray, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(img_gray, cv2.CV_64F, 0, 1, ksize=5)

# Laplaciano
laplacian = cv2.Laplacian(img_gray, cv2.CV_64F)

# Mostrar
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1), plt.imshow(sobelx, cmap='gray'), plt.title("Sobel X"), plt.axis('off')
plt.subplot(1, 3, 2), plt.imshow(sobely, cmap='gray'), plt.title("Sobel Y"), plt.axis('off')
plt.subplot(1, 3, 3), plt.imshow(laplacian, cmap='gray'), plt.title("Laplaciano"), plt.axis('off')
plt.show()
```
